```python
import pandas as pd
from pathlib import Path
import yaml

import seaborn as sns
```

```python
experiments_folder = Path("../data/experiments")
```

```python
def extract_product_params(cfg_data:dict):
    cfg_list = []
    pass

def consolidate_metrics(experiment_path:Path):
    
    product_metrics_files = experiment_path.rglob("**/metrics_products.csv")
    
    metrics_list = []
    for file_path in product_metrics_files:
        run = file_path.parent.name.split("_")[-1]
        metric_temp = pd.read_csv(file_path)
        metric_temp["run"] = run
        metrics_list.append(metric_temp)
    
    return pd.concat(metrics_list, ignore_index=True)

```

```python
metric_list = []
for experiment in experiments_folder.iterdir():
    if not experiment.is_file():
        exp_metric_df = consolidate_metrics(experiment)
        exp_metric_df["experiment"] = experiment.name
        metric_list.append(exp_metric_df)
    
metric_df = pd.concat(metric_list, ignore_index=True)
metric_df
```
